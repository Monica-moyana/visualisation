## Investigation Overview

I will be investigating  different factors that affect student exam scores. These factors are race, gender, lunch, parental level of education and test preparation course.


## Dataset Overview

This dataset is found on [Kaggle](https://www.kaggle.com/datasets/spscientist/students-performance-in-exams). It includes scores from three exams and a variety of personal, social, and economic factors that have interaction effects upon them. It is made up of 1000 rows and 8 columns

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df_student = pd.read_csv('StudentsPerformance.csv')
```

```python
#i am going to add an additional column that shows the aggregate of the learners
col = df_student.loc[: , 'math score':'writing score']
df_student['aggregate score'] = col.mean(axis=1)
```

## Performance of students by gender
The sutdents were grouped into males and females. The graph shows that there isn't a significant difference between the performance of males and females.The upper quartile, median and lower quartile are almost the same for both male and female

```python
base_color = sb.color_palette()[9]
plt.figure(figsize = (8,5))
sb.violinplot(data=df_student, x='gender', y='aggregate score', color=base_color,inner='quartile');
plt.title('performance of students by gender');
```

## Effects of taking lunch befere the exam

There are two categeries for lunch, standard lunch and free/reduced lunch. Sudents who took standard lunch otained higher aggregate scores than those who took low quality lunch


```python
# violin plot of lunch against aggregate score
plt.figure(figsize = (8,5))
sb.violinplot(data=df_student, x='lunch', y='aggregate score', color=base_color,inner='quartile');
plt.title('lunch and its effect on aggregate score');
```

## Effects of taking the test preparation course

The are two categories for this whch is those who completed the preparation course and those who did not. students who took the preparation course did exceptionally well on their exams. their aggregate scores bulge and widens mostly at 70 to 80

```python
# test preparation course against aggregate score
plt.figure(figsize = (8,5))
sb.violinplot(data=df_student, x='test preparation course', y='aggregate score', color=base_color,inner='quartile');
plt.title('Prep course and its effect on aggregate score');
```

## Change in aggregate score as a factor of lunch and test preparation course

It is clear that in both learners who took s the preparation course and those that didn't, those who ate standard lunch performed better. But also, taking standard lunch and a preperation course had an overall success rate in the exams


```python
sb.catplot(x="test preparation course", y="aggregate score",hue="lunch", kind="bar", data=df_student, palette="Blues");
plt.title('How lunch and prep course affect aggregate score');
```
